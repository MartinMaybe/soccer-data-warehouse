```python
import sqlite3
import pandas as pd 

# Connect to the SQLite database
connection = sqlite3.connect("../data/raw/database.sqlite")

# List all tables in the database
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table'", connection)
print(tables)
```

```python
# Get the number of rows in each table
table_names = tables["name"].tolist()
print(table_names)

for table in table_names:
    count = pd.read_sql(f"SELECT COUNT(*) as row_count FROM {table}", connection)
    print(f"{table}: {count['row_count'][0]} rows")
```

```python
# Get the first few rows of each table

for table in table_names:
    print(f"\n--- {table} ---")
    df = pd.read_sql(f"SELECT * FROM {table} LIMIT 3", connection)
    print(df.to_string())
```
